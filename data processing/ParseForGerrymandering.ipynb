```python
import json
```

```python
json_graph = json.load(open('utahGraph2.json'))
```

```python
adjacencies = json_graph['adjacency'] # adjacencies
```

```python
parsed_json = {} # to be outputted 
```

```python
for node in json_graph['nodes']:
  json_entry = {}
  id = node['id']

  # recom required fields
  json_entry['adjacent_nodes'] = [str(adjacency['id']) for adjacency in adjacencies[id]]
  json_entry['population'] = node['P0020001']
  json_entry['voting_history'] = 'D' if node['DEMOCRAT'] >= node['REPUBLICAN'] else 'R'
  json_entry['district'] = node['CD116']
  # total
  json_entry['black_pop'] = node['P0020006']
  json_entry['hispanic_pop'] = node['P0020002']
  json_entry['american_indian_pop'] = node['P0020007']
  json_entry['asian_pop'] = node['P0020008']
  json_entry['hawaiian_pop'] = node['P0020009']
  json_entry['other_pop'] = node['P0020010']
  # VAP
  json_entry['vap_population'] = node['P0040001']
  json_entry['vap_black'] = node['P0040006']
  json_entry['vap_hispanic'] = node['P0040002']
  json_entry['vap_american_indian'] = node['P0040007']
  json_entry['vap_asian'] = node['P0040008']
  json_entry['vap_hawaiian'] = node['P0040009']
  json_entry['vap_other'] = node['P0040010']
  # CVAP
  json_entry['cvap_population'] = node['CVAP_TOTAL']
  json_entry['cvap_black'] = node['CVAP_BLACK']
  json_entry['cvap_hispanic'] = node['CVAP_HISPANIC']
  json_entry['cvap_american_indian'] = node['CVAP_AMERICANINDIAN']
  json_entry['cvap_asian'] = node['CVAP_ASIAN']
  json_entry['cvap_hawaiian'] = node['CVAP_HAIWAIIAN'] # yes i know it's spelled wrong
  json_entry['cvap_other'] = node['CVAP_OTHER']

  # Election Data
  json_entry['democrat_voting'] = node['DEMOCRAT']
  json_entry['republican_voting'] = node['REPUBLICAN']

  # Summed up minority population
  json_entry['minority_pop'] = node['P0020006'] + node['P0020002'] + node['P0020007'] + node['P0020008'] + node['P0020009'] + node['P0020010']

  parsed_json[id] = json_entry
```

```python
with open('utahGerrymandering.json', 'w') as f:
  json.dump(parsed_json, f, ensure_ascii=False, indent=1)
```
