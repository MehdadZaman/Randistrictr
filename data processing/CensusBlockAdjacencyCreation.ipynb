```python
from gerrychain import Graph, Partition, Election
from gerrychain.updaters import Tally, cut_edges
import geopandas as gpd
import json
```

```python
df = gpd.read_file("marylandPreprocessedCensusBlocks.json")
df.to_crs(inplace=True, crs="epsg:26918")
graph = Graph.from_geodataframe(df, ignore_errors=True, reproject=True)
```

```python
print(len(graph['adjacency']))
```

Eliminates adjacencies if less than 100 feet (30.48 meters)

```python
for i, edgeList in enumerate(graph['adjacency']):
  graph['adjacency'][i] = [d for d in edgeList if d['shared_perim'] >= 30.48]

with open('marylandCensusBlockGraph.json', 'w') as f:
  json.dump(graph, f, ensure_ascii=False)
```
